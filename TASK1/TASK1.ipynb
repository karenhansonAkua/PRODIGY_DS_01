```python
year = "2023"
```

```python
population_data.describe()
```

```python
data_2023 = population_data[["Country Name" , year]].dropna()
```

```python
data_2023[year] = pd.to_numeric(data_2023[year], errors='coerce')
```

```python
top_20_data = data_2023.sort_values(by=year, ascending=False).head(20)
```

```python
plt.figure("Country")
plt.bar(top_20_data["Country Name"], top_20_data[year], color='skyblue', edgecolor='black')
plt.xlabel("Country")
plt.ylabel("Population of 2023")
plt.title("Population Distribution by Country in 2023")
plt.xticks(rotation = 45, ha = "right")
plt.grid(axis = "y", linestyle = "--", alpha = 0.7)
plt.tight_layout()
plt.show
```

```python
np.random.seed(0)
years = np.arange(2010, 2022)
age_data = np.random.normal(loc = 35, scale = 10, size = 1000)
#age_data = {year : np.random.normal(loc = 35, scale = 10, size = 1000) for year in years}
#all_ages = np.concatenate([age_data[year] for year in years])
```

```python
plt.figure(figsize = (12, 6))
plt.hist(age_data, bins = 30, color = "skyblue", edgecolor = "black")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution")
plt.grid(axis = "y", alpha = 0.75)
plt.show()
```

```python
#gender_data = {"Male" : 51, "Female" : 49}
gender_data = {year: np.random.choice(["Male", "Female"], size = 1000, p = [0.5,0.5]) for year in years}
all_genders =  np.concatenate([gender_data[year] for year in years])
gender_counts = pd.Series(all_genders).value_counts()
```

```python
plt.figure(figsize = (6, 4))
#plt.bar(gender_data.keys(), gender_data.values(), color = ["blue", "pink"])
plt.bar(gender_counts.index, gender_counts.values, color = ["blue","pink"])
plt.xlabel("Gender")
plt.ylabel("Percentage")
plt.title("Gender Distribution")
plt.show()
```
